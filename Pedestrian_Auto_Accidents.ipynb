```python
#Graphing number of accidents per year

sns.countplot(x='Crash Year', data=accidents).set_title('Year')

plt.show()
```

```python
#Graphing number of accidents per month 

fig,ax=plt.subplots(figsize=(15, 5))

sns.countplot(x='Crash Month', data=accidents, order=accidents['Crash Month'].value_counts().index, ax=ax).set_title('Month')

plt.show()
```

```python
#Graphing number of accidents per day of week

sns.countplot(x='Day of Week', data=accidents, order=accidents['Day of Week'].value_counts().index).set_title('Day of the week')

plt.show()
```

```python
#Graphing number of accidents per time of day

fig,ax=plt.subplots(figsize=(50, 15))

sns.countplot(x='Time of Day', data=accidents, order=accidents['Time of Day'].value_counts().index, ax=ax).set_title('Time of the Day')

plt.show()
```

```python
#Removing weather condition unknowns by creating new variable, then graph

accidents['Weather Conditions (2016+)'] = accidents['Weather Conditions (2016+)'].replace(['Uncoded & errors'], np.nan)

accidents_weather = accidents.dropna(subset=['Weather Conditions (2016+)'])

accidents['Weather Conditions (2016+)'] = accidents['Weather Conditions (2016+)'].replace(['Unknown'], np.nan)

accidents_weather2 = accidents.dropna(subset=['Weather Conditions (2016+)'])

fig,ax=plt.subplots(figsize=(15, 5))

sns.countplot(x='Weather Conditions (2016+)', data=accidents_weather2, ax=ax).set_title('Weather Conditions')

plt.show()
```

```python
#Pie chart for weather

x = [73, 12, 12, 2]

labels = ['Clear', 'Cloudy', 'Rain', 'Snow']

fig, ax = plt.subplots()

ax.pie(x, labels=labels)

ax.set_title('Weather')

plt.show()
```

```python
#Removing light condition unknowns by creating new variable, then graph

accidents['Lighting Conditions'] = accidents['Lighting Conditions'].replace(['Uncoded & errors'], np.nan)

accidents_light = accidents.dropna(subset=['Lighting Conditions'])

accidents['Lighting Conditions'] = accidents['Lighting Conditions'].replace(['Other / unknown'], np.nan)

accidents_light = accidents.dropna(subset=['Lighting Conditions'])

accidents_light2 = accidents.dropna(subset=['Lighting Conditions'])

accidents['Lighting Conditions'] = accidents['Lighting Conditions'].replace(['Unknown'], np.nan)

accidents_light3 = accidents.dropna(subset=['Lighting Conditions'])

fig,ax=plt.subplots(figsize=(15, 5))

sns.countplot(x='Lighting Conditions', data=accidents_light3, ax=ax).set_title('Lighting')

plt.show()
```

```python
#Pie chart for light

x = [54, 31, 11, 3, 2]

labels = ['Daylight', 'Dark lighted', 'Dark unlighted', 'Dusk', 'Dawn']

fig, ax = plt.subplots()

ax.pie(x, labels=labels)

ax.set_title('Lighting')

plt.show()
```

```python
#Removing victim gender unknowns by creating new variable, then graphing

accidents['Person Gender'] = accidents['Person Gender'].replace(['Uncoded & errors'], np.nan)

accidents_gender = accidents.dropna(subset=['Person Gender'])

sns.set(font_scale=1)

sns.countplot(x='Person Gender', data=accidents_gender).set_title('Victim Gender')

plt.show()
```

# Insights

Most accidents occurred in 2018.

October is the most common month for an accident.

Friday is the most common day for an accident.

6-7pm is the most popular day for an accident.

Most accidents occur in clear weather conditions.

More accidents occur in daylight, than any other lighting condition.

Males are more likely to be the victim of an accident.

Possible injuries is the most common injury reported.

However, fatal injuries occurred most in dark lighting conditions.
